```python
# 1st:
# Target : Crawl ur posts , then gets ur friend's main page link 

# Login FB and switch to ur main page

from selenium import webdriver
from selenium.webdriver.common.by import By
#from selenium.webdriver.common.keys import Keys
#from selenium.webdriver.support.ui import Select
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import NoAlertPresentException
from selenium.webdriver.support.ui import WebDriverWait
from bs4 import BeautifulSoup as bs
import unittest, time, re , random , os



# Define functions

# Crawling Posts :
# num = number of times for Scrolling to button

def getSourceCode(num):
    count=0
    while(count != num):
        
        # Scroll to the Button
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        # time Sleep
        time.sleep(random.randint(1,3))
        count +=1
        
    return driver.page_source


# Save Your Posts :

def savePost_yours(sourceCode):
    
    dir_name = 'FB_Users_Post_Sourcecode'
    
    if not os.path.exists(dir_name):
        os.mkdir(dir_name)

    url = '/'.join(os.getcwd()+'\{}\yourposts.html'.format(dir_name).split('\\'))
    with open(url,'w') as out_file:
        out_file.write(str(sourceCode))
    out_file.close()

    
# Save Your Friends' url :

def saveUrls_friend(friend_url_bscode):
    
    # Use Regular Expression filter friends' url  
    # Use dic to clean the repeated url
    dic = {}

    for ele in friend_url_bscode.select('a'):
        m = re.search('(.*profile.*friends_tab)',str(ele.get('href')))
        if m :
            text = m.group(1)
            if text in dic:
                dic[text] += 1
            else:
                dic[text] = 1
                
    dir_name2 = 'FB_Users_Post_Sourcecode'
                
    url2 = '/'.join(os.getcwd()+'\{}\\friends_urls.txt'.format(dir_name2).split('\\'))
    with open(url2,'w') as out_file:
        for key in dic.keys():
            out_file.write(key +'\n')
    out_file.close()
    
    
# Save Your Friend's posts:

def savePost_friend(friend_post_bscode,url,error_time):
    
    dir_name = 'FB_Friends_Post_Sourcecode'
    
    if not os.path.exists(dir_name):
        os.mkdir(dir_name)
    
    m = re.match('.*id=(\d*).*',url)
     
    if m:
        id_number = m.group(1)   
    else:
        id_number = 'Error' + str(error_time)
    location = '/'.join(os.getcwd()+'\{0}\\{1}.html'.format(dir_name,id_number).split('\\'))
   
    with open(location,'w') as out_file:
        out_file.write(str(friend_post_bscode))
    out_file.close()
    
# Selenium into FB 

account = ""
key = ""


# Open the Firefox , link to FB

driver = webdriver.Firefox()
driver.base_url = "https://www.facebook.com/"
driver.get(driver.base_url)

# Set the Static time.
# If webdriver stop action over 10sec , then show the error message

driver.implicitly_wait(10)
time.sleep(2)


# login into FB 

driver.find_element_by_id("email").send_keys(account)
driver.find_element_by_id("pass").send_keys(key)
driver.execute_script(" \
    var input_test = document.getElementsByTagName('input');\
    var button_test = document.getElementsByTagName; \
    if (input_test[3].getAttribute('value') == '登入'){ \
        input_test[3].click(); \
    } \
")

# These are failure test

#a= len(driver.find_element_by_xpath('//input[value="登入"]'))
#b= len(driver.find_element_by_xpath('//button[text()="登入"]'))
#driver.find_element_by_id('u_0_v').click()
#driver.find_element_by_id('loginbutton').click()
#driver.find_element_by_class_name('_2dpe _1ayn').click()

time.sleep(1)
print 'Logging Successful !!!'

# Set sleep time , then go to ur main page

driver.find_element_by_class_name('_2dpb').click()
#driver.find_element_by_link_text("鄭宇軒").click()

time.sleep(2)

# Get your FB Posts

urs_pageSourceCode = getSourceCode(80)
print 'Crawling Posts Successful  !!!'

# Save as File

soup = bs(urs_pageSourceCode)
savePost_yours(soup)

print 'Saving Ur posts as file Successful  !!!'
time.sleep(2)


# Go to ur friends wall , and get source code

driver.find_element_by_class_name('_5opl').click()
time.sleep(3)
friendsWall_pageSourceCode = getSourceCode(50)

print 'Crawling Friend\'s Wall Successful  !!!'

# Save friend's link as file

soup_friends = bs(friendsWall_pageSourceCode)
saveUrls_friend(soup_friends)

print 'Saving friends link file Successful  !!!'


# ---------------------------------------------------------------------------------------

# 2nd:
# Target : Crawl ur friend's posts

# Load the urls file

link = '/'.join('C:\Users\BigData\FB_Users_Post_Sourcecode\\friends_urls.txt'.split('\\'))
input_file = open(link,'r')
urls = input_file.readlines()
input_file.close()


for i in range(1,len(urls)+1):    
    url_friend = urls[i-1]
    driver.get(url_friend)
    friend_postSourceCode = getSourceCode(80)
    soup_friend = bs(friend_postSourceCode)
    savePost_friend(soup_friend,url_friend,i)
    
    print str(i) + 'times Successful !!!!'

print 'Saving friends post as file Successful  !!!' 

print "The End"
```
